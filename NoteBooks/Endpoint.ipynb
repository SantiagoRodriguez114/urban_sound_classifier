```python
from sagemaker.tensorflow import TensorFlowModel
from sagemaker import get_execution_role
import sagemaker, numpy as np
```

```python
sess = sagemaker.Session()
role = get_execution_role()
model_s3_uri = "s3://parcial-3urbansound/model.tar.gz"
```

```python
tf_model = TensorFlowModel(
    model_data=model_s3_uri,
    role=role,
    framework_version="2.13",
)
```

```python
endpoint_name = f"urbansound-{int(np.random.randint(1e9))}"
predictor = tf_model.deploy(
    initial_instance_count=1,
    instance_type="ml.c5.xlarge",
    endpoint_name=endpoint_name,
    model_data_download_timeout=900,
    container_startup_health_check_timeout=900,
)

print("Endpoint listo:", predictor.endpoint_name)
```
